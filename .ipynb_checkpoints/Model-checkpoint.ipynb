# Machine Learning models 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from imblearn.over_sampling import SMOTE
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_curve, roc_auc_score
```

```python
filePath = './data/breast_cancer_data_cleaned.xlsx'

# Load the data from the Excel file
df = pd.read_excel(filePath)

# Separate the features and the target variable
X = df.drop(columns=['cancer_recurrence'])
Y = df['cancer_recurrence']

# Print the number of columns/features
print(f"Number of features: {X.shape[1]}")

# Print the number of instances/rows
print(f"Number of instances: {X.shape[0]}")
```

Balancing dataset

```python
# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Plot the distribution of the training set before and after SMOTE
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.countplot(x=Y_train, ax=axes[0])
axes[0].set_title('Distribution of Training Set')
axes[0].set_xlabel('Cancer Recurrence')
axes[0].set_ylabel('Count')

sns.countplot(x=Y_test, ax=axes[1])
axes[1].set_title('Distribution of Test Set')
axes[1].set_xlabel('Cancer Recurrence')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Initialize SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE to the training data
X_train_balanced, Y_train_balanced = smote.fit_resample(X_train, Y_train)

# Print the number of instances in the balanced dataset
print(f"Number of instances in the balanced dataset: {X_train_balanced.shape[0]}")

# Plot the distribution of the balanced training set
plt.figure(figsize=(6, 5))
sns.countplot(x=Y_train_balanced)
plt.title('Distribution of Balanced Training Set')
plt.xlabel('Cancer Recurrence')
plt.ylabel('Count')
plt.show()
```

Logistic regression 

```python
# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Create and train the logistic regression model
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, Y_train)

# Predict the target variable for the test set
Y_pred = log_reg.predict(X_test)

# Create the confusion matrix
cm = confusion_matrix(Y_test, Y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=log_reg.classes_)
disp.plot()

plt.show()

# Calculate evaluation metrics
accuracy = accuracy_score(Y_test, Y_pred)
precision = precision_score(Y_test, Y_pred)
recall = recall_score(Y_test, Y_pred)
auc_logreg = roc_auc_score(Y_test, Y_pred)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"AUC: {auc}")

# Calculate ROC curve
fpr_logreg, tpr_logreg, _ = roc_curve(Y_test, Y_pred)

# Plot ROC curve
plt.figure()
plt.plot(fpr_logreg, tpr_logreg, color='darkorange', lw=2, label=f'ROC curve (area = {auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

Decision trees 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.tree import plot_tree

# Create and train the decision tree model
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train, Y_train)

# Predict the target variable for the test set
Y_pred_tree = decision_tree.predict(X_test)
Y_prob_tree = decision_tree.predict_proba(X_test)[:, 1]

# Calculate ROC curve and AUC for the decision tree
fpr_tree, tpr_tree, _ = roc_curve(Y_test, Y_prob_tree)
auc_tree = roc_auc_score(Y_test, Y_prob_tree)

# Plot ROC curves
plt.figure()
plt.plot(fpr_logreg, tpr_logreg, color='darkorange', lw=2, label=f'Logistic Regression (AUC = {auc:.2f})')
plt.plot(fpr_tree, tpr_tree, color='blue', lw=2, label=f'Decision Tree (AUC = {auc_tree:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()

# Plot the decision tree
plt.figure(figsize=(20, 10))
plot_tree(decision_tree, feature_names=X.columns.tolist(), class_names=['No Recurrence', 'Recurrence'], filled=True)
plt.title('Decision Tree Visualization')
plt.show()
```

Support Vector Machine

```python
from sklearn.svm import SVC

# Create and train the SVM model
svm_model = SVC(probability=True, random_state=42)
svm_model.fit(X_train, Y_train)

# Predict the target variable for the test set
Y_pred_svm = svm_model.predict(X_test)
Y_prob_svm = svm_model.predict_proba(X_test)[:, 1]

# Calculate ROC curve and AUC for the SVM
fpr_svm, tpr_svm, _ = roc_curve(Y_test, Y_prob_svm)
auc_svm = roc_auc_score(Y_test, Y_prob_svm)

# Plot ROC curves
plt.figure()
plt.plot(fpr_logreg, tpr_logreg, color='darkorange', lw=2, label=f'Logistic Regression (AUC = {auc_logreg:.2f})')
plt.plot(fpr_tree, tpr_tree, color='blue', lw=2, label=f'Decision Tree (AUC = {auc_tree:.2f})')
plt.plot(fpr_svm, tpr_svm, color='green', lw=2, label=f'SVM (AUC = {auc_svm:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

# Create and train the Naive Bayes model
nb_model = GaussianNB()
nb_model.fit(X_train, Y_train)

# Predict the target variable for the test set
Y_pred_nb = nb_model.predict(X_test)
Y_prob_nb = nb_model.predict_proba(X_test)[:, 1]

# Calculate ROC curve and AUC for the Naive Bayes model
fpr_nb, tpr_nb, _ = roc_curve(Y_test, Y_prob_nb)
auc_nb = roc_auc_score(Y_test, Y_prob_nb)

# Plot ROC curves
plt.figure()
plt.plot(fpr_logreg, tpr_logreg, color='darkorange', lw=2, label=f'Logistic Regression (AUC = {auc_logreg:.2f})')
plt.plot(fpr_tree, tpr_tree, color='blue', lw=2, label=f'Decision Tree (AUC = {auc_tree:.2f})')
plt.plot(fpr_svm, tpr_svm, color='green', lw=2, label=f'SVM (AUC = {auc_svm:.2f})')
plt.plot(fpr_nb, tpr_nb, color='red', lw=2, label=f'Naive Bayes (AUC = {auc_nb:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

Multiple linear regression 

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Create and train the multiple linear regression model
linear_reg = LinearRegression()
linear_reg.fit(X_train, Y_train)

# Predict the target variable for the test set
Y_pred_linear = linear_reg.predict(X_test)

# Calculate evaluation metrics
mse = mean_squared_error(Y_test, Y_pred_linear)
r2 = r2_score(Y_test, Y_pred_linear)

print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")

# Plot the predicted probabilities vs actual values
plt.figure(figsize=(10, 6))
plt.scatter(Y_test, Y_pred_linear, color='blue')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Probabilities')
plt.title('Multiple Linear Regression: Actual vs Predicted')
plt.show()
```
