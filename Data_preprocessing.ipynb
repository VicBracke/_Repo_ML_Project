```python
from ucimlrepo import fetch_ucirepo 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
# fetch dataset 
breast_cancer = fetch_ucirepo(id=14) 
  
# data (as pandas dataframes) 
X = breast_cancer.data.features 
y = breast_cancer.data.targets 
  
# metadata 
print(breast_cancer.metadata) 
  
# variable information 
print(breast_cancer.variables) 

# Combine features and target into a single DataFrame
df = pd.concat([X, y], axis=1)

# Print the number of columns/features
print(f"Number of features: {df.shape[1] - 1}")

# Print the number of instances/rows
print(f"Number of instances: {df.shape[0]}")

# Save the DataFrame to a CSV file
# df.to_csv('data/breast_cancer_data.csv', index=False)

display(df)
```

Look closely at variables and change wrong names variables

```python
for column in df.columns:
    unique_values = df[column].unique()
    print(f"Unique values for {column}: {unique_values}")
    print(df[column].value_counts())
    print('-----------------------')
```

```python
# Replace the values in the 'tumor-size' column
df['tumor-size'] = df['tumor-size'].replace({'14-Oct': '10-14', '9-May': '5-9'})

# Replace the values in the 'inv-nodes' column
df['inv-nodes'] = df['inv-nodes'].replace({'8-Jun': '6-8', '11-Sep': '9-11', '5-Mar': '3-5', '14-Dec': '12-14'})

# Display the updated DataFrame to verify the changes
display(df)
```

Dealing with missing values

```python
missing_values = df.isnull().sum()
missing_columns = missing_values[missing_values > 0]
print("Columns with missing values:")
print(missing_columns)

df_complete = df.dropna()
print(f"Number of instances after removing missing values: {df_complete.shape[0]}")

# Check whether there are no missing values in df_complete
no_missing_values = df_complete.isnull().sum().sum() == 0
print(f"No missing values in df_complete: {no_missing_values}")

display(df_complete)

```

hot encoding variables


```python
# Copy the original DataFrame
df_new = df_complete.copy()

# Encode 'Class' variable
df_new['cancer_recurrence'] = df_new['Class'].map({'no-recurrence-events': 0, 'recurrence-events': 1})

# Encode 'irradiat' variable
df_new['radiotherapy'] = df_new['irradiat'].map({'no': 0, 'yes': 1})

# Encode 'breast' variable
df_new['left_breast'] = df_new['breast'].map({'left': 1, 'right': 0})

# Encode 'node-caps' variable
df_new['cancer_LN_capsule'] = df_new['node-caps'].map({'no': 0, 'yes': 1})

# Encode 'menopause' variable
df_new['post_menopause'] = df_new['menopause'].map({'premeno': 0, 'lt40': 1, 'ge40': 1})

# Copy 'deg-malig' column and rename it to 'tumor_grade'
df_new['tumor_grade'] = df_new['deg-malig']

# Encode 'inv-nodes' variable
inv_nodes_mapping = {
    '0-2': 0, '3-5': 1, '6-8': 2, '9-11': 3, '12-14': 4,
    '15-17': 5, '18-20': 6, '21-23': 7, '24-26': 8, '27-29': 9,
    '30-32': 10, '33-35': 11, '36-39': 12
}
df_new['amount_cancer_LN'] = df_new['inv-nodes'].map(inv_nodes_mapping)

tumor_size_mapping = {
    '0-4': 0, '5-9': 1, '10-14': 2, '15-19': 3, '20-24': 4, '25-29': 5, 
    '30-34': 6, '35-39': 7, '40-44': 8, '45-49': 9, '50-54': 10, '55-59': 11
}
df_new['tumor_size'] = df_new['tumor-size'].map(tumor_size_mapping)

age_mapping = {
    '10-19': 0, '20-29': 1, '30-39': 2, '40-49': 3, '50-59': 4, 
    '60-69': 5, '70-79': 6, '80-89': 7, '90-99': 8
}
df_new['age_patient'] = df_new['age'].map(age_mapping)

# Hot encode the 'breast-quad' variable with prefix 'affected' and convert boolean to int
df_new = pd.get_dummies(df_new, columns=['breast-quad'], prefix='affected')
df_new = df_new.astype({col: int for col in df_new.columns if col.startswith('affected')})

# Drop the specified columns
df_new.drop(columns=['age', 'menopause', 'tumor-size', 'inv-nodes', 'node-caps', 'deg-malig', 'breast', 'irradiat', 'Class'], inplace=True)

# Display the first few rows to verify the encoding
display(df_new)

# Print the number of columns/features (excluding the output variable)
print(f"Number of features: {df_new.shape[1] - 1}")

# Print the number of instances/rows
print(f"Number of instances: {df_new.shape[0]}")
```
